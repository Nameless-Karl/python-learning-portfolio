```python
def is_year_leap(year):
    # A year is a leap year if it meets the following conditions:
    #  It is evenly divisible by 4.
    # EXCEPTION 1: If it is also evenly divisible by 100, it is not a leap year.
    # EXCEPTION 2: Unless it is also evenly divisible by 400, in which case it is a leap year. 
    
    # Write your code here.
    while year % 4 == 0:    # Every leap year must be divisible by 4
        leap_year = True
        if leap_year:   # If the year is divisible 4
            if year % 100 != 0: # The year should not be divisible by 100
                leap_year = True
                return leap_year
            elif year % 400 == 0: # The year should be divisible by 400
                leap_year = True
                return leap_year
            else:
                leap_year = False # This cover others that are divisible by 4 but not a leap year
                return leap_year
    else:
        leap_year = False
            
        return leap_year
```

```python
# Test code

test_data = [1900, 2000, 2016, 1987] # List of years
test_results = [False, True, True, False] # List of answers if year is leap or not
for i in range(len(test_data)): # For loop with the list of years as range
    yr = test_data[i]    # Assigning value to variable (yr) as index call of list of years
    print(yr,"->",end="")
    result = is_year_leap(yr)     # Assigning the of function of years to var (result)
    if result == test_results[i]:  # Comparing the results of function call to answer
        print("OK")      # Parsing results
    else:
        print("Failed")
```
