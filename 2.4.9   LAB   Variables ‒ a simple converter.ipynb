```python
kilometers = float(input("What is the distance? \n"))
miles = float(input("Going american' heey?\nHow far is the bungee? \n\n"))

miles_to_kilometers = 1.61 * miles
kilometers_to_miles = 0.6211 * kilometers

print(miles, "miles is", round(miles_to_kilometers, 2), "kilometers\n\n")
print(kilometers, "kilometers is", round(kilometers_to_miles, 2), "miles")
```

```python
kilometers = 12.25
miles = 7.38

miles_to_kilometers = miles * 1.61
kilometers_to_miles = kilometers / 1.61

print(miles, "miles is", round(miles_to_kilometers, 2), "kilometers")
print(kilometers, "kilometers is", round(kilometers_to_miles, 2), "miles")
```
