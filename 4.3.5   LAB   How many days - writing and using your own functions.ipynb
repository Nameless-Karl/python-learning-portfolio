```python
def is_year_leap(year):
    #
    # Your code from the previous LAB.
    #
    if year%4 != 0:
        outcome = False
    elif year%100 != 0:
        outcome = True
    elif year%400 != 0:
        outcome = False
    else:
        outcome = True
    return outcome
```

```python
def days_in_month(year, month):
    
    month -= 1
    days =[31, 28, 31,30,31,30,31,31,30,31,30,31]
    m = days[month]
    
    check = is_year_leap(year)
    
    if m == 28:
        if check == True:
            m = 29
        else:
            m = 28
    return m
```

```python
test_years = [1900, 2000, 2016, 1987]
test_months = [2, 2, 1, 11]
test_results = [28, 29, 31, 30]
for i in range(len(test_years)):
    yr = test_years[i]
    mo = test_months[i]
    print(yr, mo, "->", end="")
    result = days_in_month(yr, mo)
    if result == test_results[i]:
        print("OK")
    else:
        print("Failed")
```
